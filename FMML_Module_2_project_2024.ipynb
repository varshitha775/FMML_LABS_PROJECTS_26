We will be analysing the dataset using the fact that weather the drink comes under the category tea or not

````python
data['Beverage_category'].unique()
````

````python
data['Tea'] = data['Beverage_category'].apply(lambda x: 1 if x == 'Tazo Tea Drinks' else 0)
data = data.drop('Beverage_category', axis=1)
````

````python
##  one hot encoding of categorical features in data
def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix=prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
````

````python
data = onehot_encode(
    data,
    columns=['Beverage', 'Beverage_prep'],
    prefixes=['bev', 'bevp']
)
````

````python

data = data.replace({True: 1, False: 0})
````

````python
data = data.applymap(lambda x: np.float64(str(x).replace('%', '')))
````

````python
# prompt: ## to do
# # Create a pie chart of the 'Tea' column also write your observation form the plot

# Create a pie chart of the 'Tea' column
plt.pie(data['Tea'].value_counts(), labels=['Non-Tea', 'Tea'], autopct='%1.1f%%')
plt.title('Percentage of Tea and Non-Tea Drinks')
plt.show()

# Observation:
# Based on the pie chart, we can observe the proportion of Tea drinks (Beverage_category = 'Tazo Tea Drinks')
# compared to other drink categories in the dataset. This can help understand the prevalence of tea-based
# items in the Starbucks menu.
````

````python
# prompt: ## to do
# # perform pca on the data and plot the explained variace ratio, what is the optimal number of principal components in this case ?

from sklearn.preprocessing import StandardScaler

# Extract numeric features for PCA
numeric_data_starbucks = data.select_dtypes(include=['number'])

# Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data_starbucks)

# Perform PCA
pca_starbucks = PCA()
pca_starbucks.fit(scaled_data)

# Plot explained variance ratio
plt.plot(np.cumsum(pca_starbucks.explained_variance_ratio_), marker='o')
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.title('Explained Variance Ratio for Starbucks Data')
plt.show()

# Determine the optimal number of components
# You can choose a threshold (e.g., 95%) for cumulative explained variance
# or use the elbow method visually to find the point where adding more components
# doesn't significantly improve the explained variance.

# For example, if you want to retain 95% of the variance, you could find the number of
# components where the cumulative explained variance crosses 0.95:
explained_variance_threshold = 0.95
n_components_starbucks = np.argmax(np.cumsum(pca_starbucks.explained_variance_ratio_) >= explained_variance_threshold) + 1

print(f"Optimal number of principal components to retain {explained_variance_threshold * 100}% variance: {n_components_starbucks}")
````

````python
# prompt: ## to do
# # visualise the principal components, choose the number of principal components based on the above plot. What is you observation from the plot?

# Perform PCA with the chosen number of components
pca_starbucks = PCA(n_components=n_components_starbucks)
pca_starbucks.fit(scaled_data)
pca_data_starbucks = pca_starbucks.transform(scaled_data)

# Create a DataFrame with the principal components and Tea labels
pca_df_starbucks = pd.DataFrame(
    data=pca_data_starbucks,
    columns=[f"PC{i+1}" for i in range(n_components_starbucks)]
)
pca_df_starbucks["Tea"] = data["Tea"]

# Visualize the PCA results with a scatter plot (for the first two principal components)
if n_components_starbucks >= 2:
  sns.scatterplot(x="PC1", y="PC2", hue="Tea", data=pca_df_starbucks)
  plt.title("PCA Visualization of Starbucks Data")
  plt.show()

# Observation:
# Based on the scatter plot of the first two principal components, you can see how well the data is separated based on whether a drink is considered a 'Tea' drink or not.
# If the data is well-separated, it suggests that the selected features contain information related to the Tea/Non-Tea categorization. If the data points overlap significantly, it may mean that the PCA components don't provide much separation, indicating that the features aren't strong discriminators between Tea and Non-Tea drinks.
````

````python
# prompt: ## to do
# # plot the first 2 components of tsne, whats you observation from the plot?

# Initialize and fit the TSNE model
tsne_starbucks = TSNE(n_components=2)
tsne_data_starbucks = tsne_starbucks.fit_transform(scaled_data)

# Create a DataFrame with the TSNE components and Tea labels
tsne_df_starbucks = pd.DataFrame({
    "tsne_1": tsne_data_starbucks[:, 0],
    "tsne_2": tsne_data_starbucks[:, 1],
    "Tea": data["Tea"]
})

# Visualize the TSNE results with a scatter plot
sns.scatterplot(x="tsne_1", y="tsne_2", hue="Tea", data=tsne_df_starbucks)
plt.title("TSNE Visualization of Starbucks Data")
plt.show()

# Observation:
# Based on the TSNE plot, you can analyze the separation of the data based on whether a drink is a 'Tea' drink or not.
# If the clusters are distinct, it indicates that the TSNE transformation has effectively captured the underlying structure in the data,
# which can differentiate between Tea and Non-Tea categories. If the data points overlap significantly, it may suggest that the chosen features
# do not provide strong enough information to distinguish between Tea and Non-Tea categories effectively.
````
