### Gradient Descent

````python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
# Superimpose the gradient descent calculations on plot

# Gradient Descent for Task 1

new_x = 3                # Starting point
previous_x = 0           # Previous point
step_multiplier = 0.1    # Learning rate
precision = 0.00001      # Precision for stopping criterion

x_list = [new_x]         # To store the values of x
slope_list = [df(new_x)] # To store the slopes (derivative) at each x value

# Perform Gradient Descent
for n in range(500):  # max 500 iterations
    previous_x = new_x

    # Calculate gradient (derivative of the function)
    gradient = df(new_x)  # derivative of the function

    # Update x value (moving in the opposite direction of the gradient)
    new_x = new_x - step_multiplier * gradient  # Gradient descent update rule

    # Calculate the step size to check if we're converging
    step_size = abs(new_x - previous_x)

    # Store the new x and its derivative
    x_list.append(new_x)
    slope_list.append(df(new_x))

    # Stop the loop if the step size is smaller than the desired precision
    if step_size < precision:
        print('Loop ran this many times:', n)
        break

# Print results
print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))

# Plot the results
plt.figure(figsize=[20, 5])

# 1st Chart: Cost Function f(x)
plt.subplot(1, 3, 1)
plt.xlim(-3, 3)
plt.ylim(0, 8)
plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.8)
values = np.array(x_list)
plt.scatter(x_list, f(values), color='red', s=100, alpha=0.6)

# 2nd Chart: Derivative f'(x)
plt.subplot(1, 3, 2)
plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)
plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

# 3rd Chart: Derivative (Close-up)
plt.subplot(1, 3, 3)
plt.title('Gradient Descent (close up)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.grid()
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)
plt.plot(x_1, df(x_1), color='skyblue', linewidth=6, alpha=0.8)
plt.scatter(x_list, slope_list, color='red', s=300, alpha=0.6)

plt.show()
````

## Gradient Descent

````python

## Ensure this is your gradient descent function:
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001, max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list

# Example derivative function for g(x) = x^2
def g(x):
    return x**2

def dg(x):
    return 2*x

# Call gradient descent with a starting guess and get results
initial_guess = 2
final_x, x_list, slopes = gradient_descent(dg, initial_guess)

# Now, plot the results
plt.figure(figsize=[12, 6])

# Plot the x values during the descent
plt.subplot(1, 2, 1)
plt.plot(x_list, label='x values')
plt.axhline(0, color='black', linewidth=1)
plt.title('Gradient Descent: x values over iterations')
plt.xlabel('Iterations')
plt.ylabel('x values')
plt.legend()

# Plot the slopes during the descent
plt.subplot(1, 2, 2)
plt.plot(slopes, label="Slope (derivative)", color='r')
plt.axhline(0, color='black', linewidth=1)
plt.title('Gradient Descent: Derivative (Slope) over iterations')
plt.xlabel('Iterations')
plt.ylabel('Slope')
plt.legend()

plt.show()
````

Minimise $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

````python
## TASK-8 : Code the same things as they were done for the first two examples
import numpy as np
import matplotlib.pyplot as plt

# Define a new function h(x) and its derivative dh(x)
def h(x):
    return x**2 - 4*x + 4  # A simple quadratic function: f(x) = x^2 - 4x + 4

def dh(x):
    return 2*x - 4  # Derivative of the quadratic function: f'(x) = 2x - 4

# Gradient Descent function (with cost tracking)
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001, max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]
    cost_list = [h(new_x)]  # Track the cost (h(x)) at each iteration

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        cost_list.append(h(new_x))  # Append the cost at this iteration

        if step_size < precision:
            break

    return new_x, x_list, slope_list, cost_list

# Number of iterations for the experiment
n = 100

# Run Gradient Descent for different learning rates (low, mid, high, insane)
low_gamma = gradient_descent(derivative_func=dh, initial_guess=5, multiplier=0.0005, precision=0.0001, max_iter=n)
mid_gamma = gradient_descent(derivative_func=dh, initial_guess=5, multiplier=0.001, precision=0.0001, max_iter=n)
high_gamma = gradient_descent(derivative_func=dh, initial_guess=5, multiplier=0.002, precision=0.0001, max_iter=n)
insane_gamma = gradient_descent(derivative_func=dh, initial_guess=5, multiplier=0.25, precision=0.0001, max_iter=n)

# Plot the reduction in cost for each learning rate
plt.figure(figsize=[15, 10])

# Plot for low_gamma
plt.subplot(2, 2, 1)
plt.plot(low_gamma[3], label='Cost (low gamma)', color='blue')
plt.title('Low Gamma (0.0005): Cost over iterations')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.legend()

# Plot for mid_gamma
plt.subplot(2, 2, 2)
plt.plot(mid_gamma[3], label='Cost (mid gamma)', color='green')
plt.title('Mid Gamma (0.001): Cost over iterations')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.legend()

# Plot for high_gamma
plt.subplot(2, 2, 3)
plt.plot(high_gamma[3], label='Cost (high gamma)', color='red')
plt.title('High Gamma (0.002): Cost over iterations')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.legend()

# Plot for insane_gamma
plt.subplot(2, 2, 4)
plt.plot(insane_gamma[3], label='Cost (insane gamma)', color='purple')
plt.title('Insane Gamma (0.25): Cost over iterations')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.legend()

plt.tight_layout()
plt.show()

# Plot x values and slopes (derivative) for each learning rate

plt.figure(figsize=[15, 10])

# Plot for low_gamma: x values over iterations
plt.subplot(2, 2, 1)
plt.plot(low_gamma[1], label='x values (low gamma)', color='blue')
plt.axhline(0, color='black', linewidth=1)
plt.title('Low Gamma (0.0005): x values over iterations')
plt.xlabel('Iterations')
plt.ylabel('x values')
plt.legend()

# Plot for mid_gamma: x values over iterations
plt.subplot(2, 2, 2)
plt.plot(mid_gamma[1], label='x values (mid gamma)', color='green')
plt.axhline(0, color='black', linewidth=1)
plt.title('Mid Gamma (0.001): x values over iterations')
plt.xlabel('Iterations')
plt.ylabel('x values')
plt.legend()

# Plot for high_gamma: x values over iterations
plt.subplot(2, 2, 3)
plt.plot(high_gamma[1], label='x values (high gamma)', color='red')
plt.axhline(0, color='black', linewidth=1)
plt.title('High Gamma (0.002): x values over iterations')
plt.xlabel('Iterations')
plt.ylabel('x values')
plt.legend()

# Plot for insane_gamma: x values over iterations
plt.subplot(2, 2, 4)
plt.plot(insane_gamma[1], label='x values (insane gamma)', color='purple')
plt.axhline(0, color='black', linewidth=1)
plt.title('Insane Gamma (0.25): x values over iterations')
plt.xlabel('Iterations')
plt.ylabel('x values')
plt.legend()

plt.tight_layout()
plt.show()

# Plot slopes (derivative) for each learning rate

plt.figure(figsize=[15, 10])

# Plot for low_gamma: slopes over iterations
plt.subplot(2, 2, 1)
plt.plot(low_gamma[2], label='Slope (low gamma)', color='blue')
plt.axhline(0, color='black', linewidth=1)
plt.title('Low Gamma (0.0005): Derivative (Slope) over iterations')
plt.xlabel('Iterations')
plt.ylabel('Slope')
plt.legend()

# Plot for mid_gamma: slopes over iterations
plt.subplot(2, 2, 2)
plt.plot(mid_gamma[2], label='Slope (mid gamma)', color='green')
plt.axhline(0, color='black', linewidth=1)
plt.title('Mid Gamma (0.001): Derivative (Slope) over iterations')
plt.xlabel('Iterations')
plt.ylabel('Slope')
plt.legend()

# Plot for high_gamma: slopes over iterations
plt.subplot(2, 2, 3)
plt.plot(high_gamma[2], label='Slope (high gamma)', color='red')
plt.axhline(0, color='black', linewidth=1)
plt.title('High Gamma (0.002): Derivative (Slope) over iterations')
plt.xlabel('Iterations')
plt.ylabel('Slope')
plt.legend()

# Plot for insane_gamma: slopes over iterations
plt.subplot(2, 2, 4)
plt.plot(insane_gamma[2], label='Slope (insane gamma)', color='purple')
plt.axhline(0, color='black', linewidth=1)
plt.title('Insane Gamma (0.25): Derivative (Slope) over iterations')
plt.xlabel('Iterations')
plt.ylabel('Slope')
plt.legend()

plt.tight_layout()
plt.show()
````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````
